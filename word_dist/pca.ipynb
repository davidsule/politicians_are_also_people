```python
## import pca
from sklearn.decomposition import PCA
import word_dist_script as ws
from matplotlib import pyplot as plt
from seaborn import pairplot
import pandas as pd
```

```python
word2vec = ws.load_word2vec_embs()
```

```python
emb = ws.get_every_embeddings(word2vec)
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(emb)
```

```python
data = pca.fit_transform(emb)
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1)
ax.scatter(data[:,0], data[:,1], c='')
for i in range(len(data)):
    ax.text(data[i,0], data[i,1], ws.ENTITY_LABELS_SPLIT[i])
```

```python
X = 0
Y = 1
Z = 2

df = pd.DataFrame(data, columns=['x', 'y', 'z'])
fig = pairplot(df)
for i in range(len(data)):
    fig.axes[Y,X].text(data[i,X],data[i,Y] , ws.ENTITY_LABELS_SPLIT[i])
    fig.axes[Z,X].text(data[i,X],data[i,Z] , ws.ENTITY_LABELS_SPLIT[i])
    
    fig.axes[X,Y].text(data[i,Y],data[i,X] , ws.ENTITY_LABELS_SPLIT[i])
    fig.axes[Z,Y].text(data[i,Y],data[i,Z] , ws.ENTITY_LABELS_SPLIT[i])

    fig.axes[X,Z].text(data[i,Z],data[i,X] , ws.ENTITY_LABELS_SPLIT[i])
    fig.axes[Y,Z].text(data[i,Z],data[i,Y] , ws.ENTITY_LABELS_SPLIT[i])
```
